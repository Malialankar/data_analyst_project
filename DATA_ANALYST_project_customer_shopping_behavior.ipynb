* CREATE A COLUMN AGE_GROUP 
- use this command to divide the column age according to the their age.
-  " df['age_group']=pd.qcut(df['age], q=4, labels = lables, duplicates='drop')"

```python
labels = ['young_adult','adult','middle_age','senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head()
```

- To select a fixed range then use 
- pd.cut(df['column_name'], bins= bins_youassign, labels=lables_youassign)
- bins divide data into ranges

```python
labels1 = ['baby','young_adult','adult','middle_age','senior']
bins = [0,7,18,26,50,100]
df['age_group1'] = pd.cut(df['age'], bins = bins , labels=labels1)
```
